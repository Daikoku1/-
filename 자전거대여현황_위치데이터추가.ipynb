```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/대여소정보.csv', encoding = 'CP949')
df1.head()
```

```python
df1 = df1[['대여소\n번호', '위도', '경도']]
df1.head()
```

```python
df2 = pd.read_csv('data/2106_대여이력.csv', encoding = 'CP949')
df2.head()
```

```python
df = df2[['대여일시', '대여 대여소번호', '반납일시', '반납대여소번호', '이용시간', '이용거리']]
df.head()
```

```python
df1.columns
```

```python
df = pd.merge(df, df1, left_on='대여 대여소번호', right_on='대여소\n번호', how='left')
df
```

```python
df.drop('대여소\n번호', axis='columns', inplace=True)

df.columns = ['대여일시', '대여 대여소번호', '반납일시', '반납대여소번호', '이용시간', '이용거리', 's위도', 's경도']
df.head()
```

```python
df = pd.merge(df, df1, left_on='반납대여소번호', right_on='대여소\n번호', how='left')
df.drop('대여소\n번호', axis='columns', inplace=True)
df.columns = ['대여일시', '대여대여소번호', '반납일시', '반납대여소번호', '이용시간', '이용거리', 's위도', 's경도', 'e위도', 'e경도']
df.head()
```

```python
df.to_csv('data/자전거대여현황_위도경도ver.csv')
```
