## 오기재된 데이터 제거

```python
df['대여 대여소번호'].unique()
```

```python
df['반납대여소번호'].unique()
# 문자열 데이터 포함됨
```

```python
# 문자열 데이터 제거
def check(x):
    try:
        return int(x)
    except: 
        return None
df['반납대여소번호'] = df.apply(lambda x : check(x['반납대여소번호']), axis = 1)
df.head()
```

```python
print(len(df))
df = df.dropna()
print(len(df))
df.head()
```

```python
df.info()
```

```python
df = pd.merge(df, df1, left_on='대여 대여소번호', right_on='대여소\n번호', how='left')
df
```

```python
df.drop('대여소\n번호', axis='columns', inplace=True)

df.columns = ['대여일시', '대여 대여소번호', '반납일시', '반납대여소번호', '이용시간', '이용거리', 's위도', 's경도']
df.head()
```

```python
df = pd.merge(df, df1, left_on='반납대여소번호', right_on='대여소\n번호', how='left')
df.drop('대여소\n번호', axis='columns', inplace=True)
df.columns = ['대여일시', '대여대여소번호', '반납일시', '반납대여소번호', '이용시간', '이용거리', 's위도', 's경도', 'e위도', 'e경도']
df.head()
```

```python
df.isnull().sum()
```

```python
dic = {i : 1 for i in df1['대여소\n번호'].unique()}
def check(li):
    ans = []
    for i in li:
        if i not in dic: ans.append(i)
    return ans
```

```python
a = check(df['대여대여소번호'].unique())
print(len(a))
a
```

```python
b = check(df['반납대여소번호'].unique())
print(len(b))
b
```
